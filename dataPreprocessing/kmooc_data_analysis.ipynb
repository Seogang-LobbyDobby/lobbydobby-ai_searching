```python
import pandas as pd
import re
```

```python
data = pd.read_csv('kmooc.csv')
```

```python
data = data.drop(data.loc[data['condition']=='종료'].index.to_list())
data = data.reset_index(drop=True)
```

```python
data['kind'] = data['kind'].str.replace('\n', '')
data['kind'] = data['kind'].str.replace(' ', '')
data['kind'] = data['kind'].str.replace('ㆍ', '')
data['kind'] = data['kind'].str.replace('과학', '')
```

```python
for i, d in enumerate(data['kind']):
    try:
        data['kind'][i] = re.sub(pattern=r'\([^)]*\)', repl='', string=d)
        print(data['kind'][i])
    except:
        print(data['kind'][i])
        continue
```

```python
for d in data['title']:
    print(d)
```

```python
for i, d in enumerate(data['title']):
    if d[0] == ' ':
        data['title'][i] = d[1:]
    print(data['title'][i])
```

```python
data['condition'] = data['condition'].str.replace('종료\(', '')
data['condition'] = data['condition'].str.replace('\)', '')
```

```python
data['title'] = data['title'] + ', ' + data['kind']
data['checking'] = data['title'] + ', ' + data['condition']

saving, dupli = [], set([])
for check in data['checking']:
    if check in saving:
        dupli.add(check)
    else:
        saving.append(check)
        
for ck in dupli:
    data = data.drop(data.loc[data['checking']==ck].index.to_list())

data = data.reset_index(drop=True)

data
```

```python
saving, dupli = [], set([])
for title in data['title']:
    if title in saving:
        dupli.add(title)
    else:
        saving.append(title)
        
for tt in data['title']:
    if tt in dupli:
        display(data[data['title']==tt])
```

```python
temp = data[['title', 'condition', 'url', 'kind']]
```
